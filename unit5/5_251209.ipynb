```python
# 4. 함수 인자로 체중과 신장을 받은 후, BMI 기준표에 따라, 마른 체형/표준/비만/고도비만을 출력하는 함수

def bmitest(KG, M):
  BMI = KG/(M**2)
  if BMI < 18.5:
    print('마른체형')
  elif 18.5 <= BMI < 25.0:
    print('표준')
  elif 25.0 <= BMI < 30.0:
    print('비만')
  elif BMI >= 30.0:
    print('고도비만')
  else:
    print('오류입니다')

bmitest(72, 1.78)
```

```python
def cal_bmi(weight, height):
  height = height * 0.01
  bmi = weight / (height * height)
  if bmi < 18.5:
    print('마른체형')
  elif 18.5 <= bmi < 25.0:
    print('표준')
  elif 25.0 <= bmi < 30.0:
    print('비만')
  else: 
    print('고도비만')

cal_bmi(64, 170)
```

```python
# 5. 사용자로부터 키, 몸무게를 입력받은 후, BMI 값과 그 값에 따른 체형 정보를 화면에 출ㅕ
# input 함수 사용, 사용자로부터 키와 몸무게 입력 받은 후 무한 루프로 체형정보 출력

while True:
  height = int(input('키'))
  weight = int(input('몸무게'))
  BMI = weight/((height*0.01)**2)
  if BMI < 18.5:
    print('BMI: %s, 마른체형' % BMI)
    print('키:', height)
    print('몸무게:', weight)
  elif 18.5 <= BMI < 25.0:
    print('BMI: %s, 표준' % BMI)
    print('키:', height)
    print('몸무게:', weight)
  elif 25.0 <= BMI < 30.0:
    print('BMI: %s, 비만' % BMI)
    print('키:', height)
    print('몸무게:', weight)
  elif BMI >= 30.0:
    print('BMI: %s, 고도비만' % BMI)
    print('키:', height)
    print('몸무게:', weight)
  else:
    ('오류입니다')
```

```python
# 8. 함수의 인자로 시작&끝을 나타내는 숫자를 받아 시작~끝까지의 모든 정수 값의 합 반환

def add_start_to_end(start, end):
  if start == end:
    print(start+end)
  elif start < end:
    for i in range(start, end+1):
      print(i)
    print(start+i+end+1)
  elif start > end:
    print("오류가 발생했습니다.")

add_start_to_end(1, 4)
```

```python
def start_to_end(start, end):
  return sum(range(start, end+1))

start_to_end(1, 10)
```

```python
def get_abbr(data_list):
  result = []
  for x in data_list:
    result.append(x[:3])
  return result

get_abbr(['Seoul', 'Daegu', 'Kwangju', 'Jeju'])
```
