```python
def cal_lower(price):
  decrement = price * 0.3
  lower_price = price - decrement
  return lower_price
```

```python
cal_lower(1000)
```

```python
def cal_upper_lower(price):
  offset = price * 0.3
  upper = price + offset
  lower = price - offset
  return (upper, lower)
```

```python
cal_upper_lower(10000)
```

```python
(upper, lower) = cal_upper_lower(10000)
upper
```
