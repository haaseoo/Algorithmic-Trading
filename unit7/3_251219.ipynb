```python
# 2

import os

def printlist(path):
  pathlist = os.listdir(path)
  f = open('/Users/anhaseo/Desktop/py/Algorithmic-Trading/unit7/flist.txt', 'wt')
  for i in pathlist:
    f.write("%s\n" % i)
  f.close()
```

os.listdir()은 파일 이름 (문자열)의 리스트를 반환   
파일 이름은 문자열이므로 %d 가 아닌 %s 사용
