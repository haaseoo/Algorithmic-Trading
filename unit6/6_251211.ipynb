```python
class Parent:
  def can_sing(self):
    print("Sing a song")
```

```python
father = Parent()
father.can_sing()
```

```python
class LuckyChild(Parent):
  pass
```

```python
child1 = LuckyChild()
child1.can_sing()
```

```python
class UnLuckyChild:
  pass
```

```python
child2 = UnLuckyChild()
child2.can_sing()
```

```python
class LuckyChild2(Parent):
  def can_dance(self):
    print("Shuffle Dance")
```

```python
child2 = LuckyChild2()
child2.can_sing()
```

```python
child2.can_dance()
```
