```python
class Account:
    num_accounts = 0
    def __init__(self, name):
        self.name = name
        Account.num_accounts += 1
    def __del__(self):
        Account.num_accounts -= 1

# 1. 객체 생성 전 확인
print(f"시작 전: {Account.num_accounts}")

# 2. 객체 생성
kim = Account("kim")
lee = Account("lee")

# 3. 결과 확인
print(f"kim의 카운트: {kim.num_accounts}")
print(f"lee의 카운트: {lee.num_accounts}")
print(f"클래스 전체 카운트: {Account.num_accounts}")
```
