```python
class BusinessCard:
  def __init__(self, name, email, addr):
    self.name = name
    self.email = email
    self.addr = addr
  def print_info(self):
    print("---------------------------")
    print("Name: ", self.name)
    print("E-mail: ", self.email)
    print("Address: ", self.addr)
    print("---------------------------")
```

```python
member1 = BusinessCard()
```

```python
member1 = BusinessCard("Kangsan Lee", "kangsan@gmail.com", "USA")
member1.print_info()
```
